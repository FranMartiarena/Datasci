```python
#a)

from scipy.stats import multivariate_normal
from sklearn.model_selection import train_test_split

x = np.vstack((w1, w2))
y = np.array([0]*10 + [1]*10)  # 0 = w1, 1 = w2

#Separo en conjuntos de entrenamiento y testeo

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y)

x_train_w1 = x_train[y_train == 0]
x_train_w2 = x_train[y_train == 1]

#Entrenar un dicotomizador gaussiano -> estimar los parametros y hacer un clasificador gaussiano.

mu_3d_w1 = x_train_w1.mean(axis=0)
cov_3d_w1 = np.cov(x_train_w1.T, ddof=0)

mu_3d_w2 = x_train_w2.mean(axis=0)
cov_3d_w2 = np.cov(x_train_w2.T, ddof=0)

#Suponiendo son w1 y w2 son las unicas clases, entonces p(w1) = p(w2) = 0.5

# Calcular verosimilitudes
rv1 = multivariate_normal(mean=mu_3d_w1, cov=cov_3d_w1)
rv2 = multivariate_normal(mean=mu_3d_w2, cov=cov_3d_w2)

likelihood_w1 = rv1.pdf(x_test) #Ac calculo para cada punto del conjunto de test su probabilidad segun los parametros estimados de la primera clase
likelihood_w2 = rv2.pdf(x_test) #Ac calculo para cada punto del conjunto de test su probabilidad segun los parametros estimados de la segunda clase

# Clasificar: mayor verosimilitud
y_pred = (likelihood_w2 > likelihood_w1).astype(int)

print(f" Prediccion: {y_pred}")
print(f" Valores reales: {y_test}")
```

```python
#Ahora lo hago con un clasificador mas artesanal usando una funcion discriminante.

from numpy.linalg import inv

def clasificar(x, mus, covarianzas):

  W0 = (-0.5)*inv(covarianzas[0])
  w0 = inv(covarianzas[0]) @ mus[0]

  W1 = (-0.5)*inv(covarianzas[1])
  w1 = inv(covarianzas[1]) @ mus[1]

  w00 = (-0.5)*(mus[0]).transpose() @ inv(covarianzas[0]) @ mus[0] - (0.5)*np.log(np.linalg.det(covarianzas[0])) + np.log(0.5)
  w10 = (-0.5)*(mus[1]).transpose() @ inv(covarianzas[1]) @ mus[1] - (0.5)*np.log(np.linalg.det(covarianzas[1])) + np.log(0.5)

  g0 = x.transpose() @ W0 @ x + w0.transpose() @ x + w00
  g1 = x.transpose() @ W1 @ x + w1.transpose() @ x + w10

  g = g0 - g1

  if g < 0:
    return 1
  else:
    return 0

pred = [clasificar(i, [mu_3d_w1, mu_3d_w2], [cov_3d_w1, cov_3d_w2]) for i in x_test]
print(pred)
print(y_test)
```
